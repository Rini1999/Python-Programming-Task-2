```python
#Question 11

n = int(input())
s = set(map(int, input().split()))
N = int(input())

for i in range(N):
    commands = list(map(str, input().split()))
    if commands[0] == "pop":
        s.pop()
    elif commands[0] == "remove":
      try:
        s.remove(int(commands[1]))
      except:
        continue
    elif commands[0] =="discard":
      try:
         s.discard(int(commands[1]))
      except:
        continue
print(sum(s))
```

```python
#Question 19

# Enter your code here. Read input from STDIN. Print output to STDOUT
def strict_superset(A):
    n = int(input())
    for i in range(0, n):
        sub_set = set(input().split())
        if len(sub_set.difference(A)):
            return False
        
    return True

A = set(input().split())
if strict_superset(A):
    print('True')
else:
    print('False')
```
